```python
import pandas as pd
import numpy as np

# load titanic and housing data from seaborn
import seaborn as sns

import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler, MinMaxScaler

from sklearn.model_selection import train_test_split

titanic = sns.load_dataset("titanic")


file_path_t = "/home/conda/AI-ML/titanic.csv"

titanic.to_csv(file_path_t, index=False)

titanic_data =  pd.read_csv("/home/conda/AI-ML/titanic.csv") # load
housing_data = pd.read_csv("/home/conda/AI-ML/housing.csv") # load

print("\n--------------------------- titanic ---------------------------------\n", titanic_data)
print("\n--------------------------- housing ---------------------------------\n", housing_data)

first_10_titanic_rows = titanic_data.head(10)
print("\n---------------------------first 10 titanic ---------------------------------\n", first_10_titanic_rows)

first_10_housing_rows = housing_data.head(10)
print("\n---------------------------first 10 housing ---------------------------------\n", first_10_housing_rows)

missing_titanic_data = titanic_data.isnull().sum()
print("\n---------------------------missing_titanic_data ---------------------------------\n", missing_titanic_data)

missing_housing_data = housing_data.isnull().sum()
print("\n---------------------------missing_housing_data ---------------------------------\n", missing_housing_data)

# handling missing values
def fill_missing_values(df, skew_threshold=1):
    df_filled = df.copy()
    for col in df.columns:
        if df[col].dtype in ['int64', 'float64']: #numerical columns
            skewness = df[col].skew()
            if skewness > skew_threshold or skewness < -skew_threshold:
                fill_value = df[col].median()
                method = "median"
            else:
                fill_value = df[col].mean()
                method = "mean"
        else: # categegorical column
            fill_value = df[col].mode()[0]
            method = "mode"
        # df_filled[col].fillna(fill_value, inplace=True)
        df_filled.fillna({col: fill_value}, inplace=True)
        print(f"Filled missing values in '{col}' using {method} ({fill_value})")
    return df_filled

titanic_filled = fill_missing_values(titanic)
housing_filled = fill_missing_values(housing_data)

print("\n--------------------------- titanic_filled ---------------------------------\n", titanic_filled)
print("\n--------------------------- housing_filled ---------------------------------\n", housing_filled)

# handle Outliers 

# example with titanic fare
sns.boxplot(x=titanic['fare'])
plt.show()

# example with 'age'
sns.boxplot(x=titanic['age'])

plt.show()

# Step 2 — Remove or Cap Outliers using IQR
# The IQR method:

# Q1 = 25th percentile

# Q3 = 75th percentile

# IQR = Q3 - Q1

# Outliers:

# Below Q1 - 1.5 × IQR

# Above Q3 + 1.5 × IQR

def cap_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Cap the outliers
    df[col] = df[col].clip(lower_bound, upper_bound)

# apply to Titanic numeric columns
numeric_titanic_cols = titanic.select_dtypes(include=['float64', 'int64']).columns
numeric_housing_cols = housing_data.select_dtypes(include=['float64', 'int64']).columns


for col in numeric_titanic_cols:
    print("\nBefore:")
    print(titanic_data[col].describe())
    sns.boxplot(x=titanic_data[col])
    plt.title(f"'{col}' - Before Capping")
    plt.show()
    
    cap_outliers(titanic_data, col)
    
    print("\nAfter:")
    print(titanic_data[col].describe())
    sns.boxplot(x=titanic_data[col])
    plt.title(f"'{col}' - After Capping")
    plt.show()

for col in numeric_housing_cols:
    print("\nBefore:")
    print(housing_data[col].describe())
    sns.boxplot(x=housing_data[col])
    plt.title(f"'{col}' - Before Capping")
    plt.show()
    
    cap_outliers(housing_data, col)
    
    print("\nAfter:")
    print(housing_data[col].describe())
    sns.boxplot(x=housing_data[col])
    plt.title(f"'{col}' - After Capping")
    plt.show()


# Sacale Feature
# 1 load data
# 2 select numeric columns
# 3 Fill NaN to avoid error during scaling
titanic_data[numeric_titanic_cols] = titanic_data[numeric_titanic_cols].fillna(titanic_data[numeric_titanic_cols].mean())
housing_data[numeric_housing_cols] = housing_data[numeric_housing_cols].fillna(housing_data[numeric_housing_cols].mean())
# StandardScaler

scaler_std = StandardScaler()
df_standard_t = pd.DataFrame(scaler_std.fit_transform(titanic_data[numeric_titanic_cols]), columns=numeric_titanic_cols)
df_standard_h = pd.DataFrame(scaler_std.fit_transform(housing_data[numeric_housing_cols]), columns=numeric_housing_cols)

print("\n--- Standard Scaled Data (first 5 rows) ---")
print(df_standard_t.head())
print(df_standard_h.head())
# 

# MinMaxScaler
scaler_mm = MinMaxScaler()
df_minmax_t = pd.DataFrame(scaler_mm.fit_transform(titanic_data[numeric_titanic_cols]), columns=numeric_titanic_cols)
df_minmax_h = pd.DataFrame(scaler_mm.fit_transform(housing_data[numeric_housing_cols]), columns=numeric_housing_cols)

print("\n--- MinMax Scaled Data (first 5 rows) ---")
print(df_minmax_t.head())
print(df_minmax_h.head())

# Train-test Split

#  we will use 80% train /20% test

# For demonstration, predict survival based on numeric feaures

X_t = titanic_data[numeric_titanic_cols]
Y_t = titanic_data["survived"] # target variable

#Train-test split (80/20)

X_train, X_test, Y_train, Y_test = train_test_split(X_t, Y_t, test_size=0.2, random_state=42)
print("\nTrain set shape:", X_train.shape)
print("Test set shape:", X_test.shape)

print("\nTrain set shape:", Y_train.shape)
print("Test set shape:", Y_test.shape)

```
