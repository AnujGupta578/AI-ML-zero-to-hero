```python
import numpy as np

# create an array of 10 random integeres between 1 and 100

random_numbers = np.random.randint(1, 101, size=10)
print("Random integers", random_numbers)

# Square each number

square_each_num = np.square(random_numbers)
print("square_each_num", square_each_num)

#  find and print the mean, max and min of the array

mean_num = np.mean(random_numbers)
max_num = np.max(random_numbers)
min_num = np.min(random_numbers)

print("mean_num", mean_num)
print("max_num", max_num)
print("min_num", min_num)

# sort an array

sort_array = np.sort(random_numbers)
print("Sorted Array", sort_array)

# filter number divisible by 3

num_divide_by_3 = random_numbers[random_numbers % 3 == 0]

print("Divisible by 3", num_divide_by_3)
```

```python
#  assignment 

import numpy as np

# change the size to 50 random numbers between 10 and 500

random_50_num = np.random.randint(10, 501, size=50)

print("50 random nums", random_50_num)

#  compute sum, median, and standard deviation

sum_value = np.sum(random_50_num)
median_value = np.median(random_50_num)
standard_dev_value = np.std(random_50_num)

print("sum_value", sum_value)
print("median_value", median_value)
print("standard_dev_value", standard_dev_value)

#  filter numbers that are -> Even and greater than 200 and Odd and less than 50.

even_num = random_50_num[random_50_num % 2 == 0]
odd_num =  random_50_num[random_50_num % 2 != 0]
even_greater_50 = even_num[even_num > 50]
odd_less_50 = odd_num[odd_num < 50]

print("even_greater_50", even_greater_50)
print("odd_less_50", odd_less_50)

# Save the main array
np.savetxt("/home/conda/AI-ML/day2_results.csv", random_50_num, delimiter=",", fmt="%d", header="RandomNumbers", comments='')

# Pad arrays so they can be combined
max_len = max(len(random_50_num), len(even_greater_50), len(odd_less_50))
pad = lambda arr: np.pad(arr, (0, max_len - len(arr)), constant_values=-1)  # -1 as placeholder for missing values

# Combine arrays and save
combined = np.column_stack((pad(random_50_num), pad(even_greater_50), pad(odd_less_50)))
np.savetxt("/home/conda/AI-ML/day2_filtered_results.csv", combined, delimiter=",", fmt="%d",
           header="AllNumbers,Even>50,Odd<50", comments='')

```
